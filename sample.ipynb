```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x,y)
model.score(x,y)
```

```python
prediction = model.predict(x)
plt.scatter(x,y)
plt.plot(x,prediction,c='red')
```
